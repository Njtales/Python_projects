## Word_count_python

```python
import os.path

def word_count(data):
    counts = dict()
    words = data.split()

    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1

    return counts

def enter_file_location():
    location = input("Enter file location (with file name and file extension) : ")

    if location == '':
        print("You entered nothing, hence using demo_text.txt")
        return default_file
    
    elif os.path.exists(location.encode('unicode_escape')) == False:
        print("\nInvalid file path! File does not exist.")
        print("Using demo text file.")
        return default_file
    else:
        
        print("\nUsing the location you entered: ",location)
        return location
    
default_file = "demo_text.txt"

file = open(enter_file_location(), "rt")
data = file.read()
words = data.split()

print('\nNumber of words and their count in text file are below : \n')
count_dict = word_count(data)

for i in count_dict:
    print(i + ' : ' + str(count_dict[i]))
```
