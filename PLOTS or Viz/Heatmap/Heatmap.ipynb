# Heatmap

a heatmap is used to understand the relationship between different features in a dataset. It represents numbers in the form of a coloured pallet such that darker shades represent a high degree of relationship between the features and the lighter shades represent a low degree of relationship between the features. Now letâ€™s see how to visualize a heatmap using Python.<br>
There are many libraries in Python to visualize a heatmap but the simplest way is by using the Seaborn library in Python. 

```python
import pandas as pd
cereals = pd.read_csv("cereal.csv")
cereals.head()
correlations = cereals.corr(method='pearson')

import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(15,8))
sns.heatmap(correlations, cmap="crest")

plt.show()
```

You can see in the figure above that it represents different colour variations. We can also plot the degree of relationship (which is between -1 and 1) on the heatmap:

```python
plt.figure(figsize=(15,8))
sns.heatmap(correlations, cmap="crest", annot=True)
plt.show()
```
