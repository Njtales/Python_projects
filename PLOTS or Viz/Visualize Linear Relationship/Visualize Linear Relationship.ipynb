## Visualize_Linear_Relationship

```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("winequality-red.csv", encoding = 'latin1')
data = data.dropna()
data.head()
```

```python
data.corr()
```

```python
figure = px.scatter(data_frame = data, 
                    x="fixed acidity",
                    y="density", 
                    size="quality", 
                    trendline="ols", 
                    title = "Relationship Between fixed acidity and density")
figure.show()
```

```python
plt.figure(figsize=(8, 6))
plt.style.use('fivethirtyeight')
plt.title("Relationship Bewtween fixed acidity & density")
sns.regplot(x="fixed acidity", y="density", data=data)
plt.show()
```
